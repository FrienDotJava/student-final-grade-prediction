# Exploratory Data Analysis (EDA)

```python
categorical_color = sns.color_palette("Set2")
numerical_color = sns.color_palette("mako", 8)
```

## Feature distribution

### All features

```python
# Count number of features
num_vars = df.shape[1]
 
# Determine number of rows and columns
n_cols = 4
n_rows = -(-num_vars // n_cols) 
 
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))
 
# Flatten axes array to ease iteration if needed
axes = axes.flatten()
 
# Plot every features
for i, column in enumerate(df.columns):
    df[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
 
# Delete unused subplot, if exist
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
 
plt.tight_layout()
plt.show()
```

```python
labels_bad_good = ['Very Bad', 'Bad', 'Normal', 'Good', 'Very Good']
bad_good_order = {1: 'Very Bad', 2: 'Bad', 3: 'Normal', 4: 'Good', 5: 'Very Good'}

labels_low_high = ['Very Low', 'Low', 'Normal', 'High', 'Very High']
low_high_order = {1: 'Very Low', 2: 'Low', 3: 'Normal', 4: 'High', 5: 'Very High'}

study_labels = ['<2 hours', '2 to 5 hours', '5 to 10 hours', '>10 hours']
study_order= {1: '<2 hours', 2: '2 to 5 hours', 3: '5 to 10 hours', 4: '>10 hours'}
df_cat = df.copy()

df_cat['health'] = pd.Categorical(
    df_cat['health'].map(bad_good_order),
    categories=labels_bad_good,
    ordered=True
)
df_cat['Dalc'] = pd.Categorical(
    df_cat['Dalc'].map(low_high_order),
    categories=labels_low_high,
    ordered=True
)
df_cat['Walc'] = pd.Categorical(
    df_cat['Walc'].map(low_high_order),
    categories=labels_low_high,
    ordered=True
)
df_cat['goout'] = pd.Categorical(
    df_cat['goout'].map(low_high_order),
    categories=labels_low_high,
    ordered=True
)
df_cat['freetime'] = pd.Categorical(
    df_cat['freetime'].map(low_high_order),
    categories=labels_low_high,
    ordered=True
)
df_cat['studytime'] = pd.Categorical(
    df_cat['studytime'].map(study_order),
    categories=study_labels,
    ordered=True
)

df_cat
```
